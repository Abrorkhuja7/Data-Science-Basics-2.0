## Load file and set index

```python
import pandas as pd

df = pd.read_csv("../data/auto.csv", index_col="ID")
df
```

## Count observations

```python
df.count()
```

## Drop duplicates

```python
df = df.drop_duplicates(
    subset=["CarNumber", "Make_n_model", "Fines"],
    keep="last"
)

df.count()
```

## Work with missing values

```python
df.isna().sum()
```

### - Drop columns with more than 500 missing values

```python
df = df.dropna(axis=1, thresh=len(df) - 500)

df.isna().sum()
```

### - Fill Refund with previous value

```python
df["Refund"] = df["Refund"].ffill()
df.isna().sum()
```

### - Fill Fines with mean

```python
df["Fines"] = df["Fines"].fillna(df["Fines"].mean())
df.isna().sum()
```

## Split Make and Model

```python
df["Make"] = df["Make_n_model"].apply(lambda x: str(x).split()[0])
df["Model"] = df["Make_n_model"].apply(
    lambda x: " ".join(str(x).split()[1:])
)

df = df.drop(columns=["Make_n_model"])
```

### - Save to JSON

```python
df.to_json("../data/auto.json", orient="records")
```
